### **Lambda** (교재 694p)
---
#### **1. 특징**
  - 익명함수를 간결하게 표현할 수 있는 방법
  - 프로그래밍 스타일 지원 ➡ 코드의 가독성 및 간결성을 높임

#### **2. 기본 구문**
```
LambdaEx.excuteOperation(parameters, (variables) -> {expressions};);
```
  - 변수가 없거나 2개 이상인 경우 () 생략 불가
  - 표현식이 2개 이상인 경우 {} 생략 불가
    - 무조건 {} 안에서만 세미콜론(;) 을 사용할 수 있다.

#### **3. *@FunctionalInterface***
  - 함수형 인터페이스 선언
  - 추상 메소드는 하나만 가질 수 있음
  - 컴파일러과정에서 추상메소드가 하나인지 검사

#### **4. 메소드의 참조**
- 목적
  - 메소드를 참조해 매개변수의 정보 및 리턴 타입을 알아냄
  - ➡ 람다식에서 불필요한 매개변수 제거

- 조건
  - 참조하는 메소드의 형태가 함수형 인터페이스에 선언한 추상메소드의 형태와 동일해야 함(리턴타입, 매개변수의 순서, 개수, 타입)

- 형식
  - static 메소드를 참조할 때
  ```
  Class :: Method
  ```
  - 일반(인스턴스) 메소드를 참조할 때
    - 메소드를 사용할 객체를 먼저 생성 후
    ```
    Reference :: Method
    ```

##### **4-1. 생성자 참조**
- 조건
  - 참조하는 메소드의 형태가 함수형 인터페이스에 선언한 추상메소드의 형태와 동일해야 함(리턴타입, 매개변수의 순서, 개수, 타입)
- 형식
  ```
  Class :: new
  ```


  

### **Stream** (교재 722p)
---
#### **1. 특징**
- 컬렉션 및 배열 기반
- 내부 반복
  - 처리 속도가 빠르고 병렬 처리에 효율적
  - 병렬 처리:
- 지연 평가
  - 필요한 시점에만 요소를 처리함
- 함수형 프로그래밍 스타일
  - forEach
    - 람다식으로 다양한 요소 처리를 저장할 수 있으며, 중간 처리와 최종 처리 수행을 위한 파이프라인을 형성
      - 중간 처리
        - 리턴 타입 Stream
        - 여러번 사용할 수 있음
      - 최종 처리
        - 리턴 타입 Stream 아님
        - 마지막에 한 번만 사용할 수 있음
  - 요소 필터링, 변환 등의 연산을 함수 형태로 제공
  - ➡ 메소드 호출 시, 람다식으로 함수형 인터페이스의 구현체 전달

#### **2. 기본 구문**
- 컬렉션
```
List<타입> 리스트명 = Array.asList(변수);
Stream<타입> 스트림명 = 리스트명.stream();
```
- 배열
```
타입[] 배열명 = {배열 값};
Stream<타입> 스트림명 = Arrays.stream(배열);
```
  - 기본 타입의 배열들을 스트림으로 변환하여 사용할 수 있는 기본 타입 스트림 존재
  - IntStream, LongStream, DoubleStream ...

#### **3. 응용 함수**
  - forEach()
    - Stream 요소에 하나씩 접근하여 작성한 코드 실행
    - Stream에 담겨 있는 요소의 개수만큼 코드 실행
  - map()
    - 실행 코드의 결과 값들을 모아서 새로운 stream 리턴
		- 새로운 Stream에 담길 요소들을 리턴해야 하기에 리턴 구문이 필수적
    - mapToInt()
      - Stream<제네릭> 타입의 스트림을 기본 타입 스트림인 IntStream으로 변환
      - 람다식에서 리턴되는 결과값으로 스트림 구성
    - flatMap()
      -
  - sum()
    - 기본 타입 스트림에서만 사용 가능
    - 내부의 모든 값들을 합산함
  - average()
    - 기본 타입 스트림에서만 사용 가능
    - 내부의 모든 값들의 평균을 구함
  - filter()
    - true나 false가 리턴되도록 논리식이나 조건식을 사용
  - reduce()
    ```
    reduce(누적 값의 초기값, (누적값이 될 변수, 요소에 하나씩 접근할 변수, 요소의 인덱스) -> 누적값에 누적할 실행코드);
    ```
    - 인덱스는 생략 가능
  - distinct()
    - 중복된 값을 제거한 새로운 Stream으로 리턴
    - 객체의 경우, 주소값이 같으면 중복으로 인식
  - peek()
    - forEach와 동일한 동작하지만, 원본 스트림이 그대로 리턴하는 중간 단계 메소드임
    - ➡ 최종 단계 메소드가 없으면 동작하지 않음.
    - 중간에 Stream의 요소값을 확인할 때 사용
  - Match
    - allMatch(람다식)
      - 람다식의 조건식에 모든 요소가 true여야 true 리턴
      - &&
    - anyMatch(람다식)
      - 람다식의 조건식에 하나의 요소라도 부합하면 true 리턴
      - ||
    - noneMatch(람다식)
      - 람다식의 조건식에 모든 요소가 부합하지 않아야 true 리턴
  - count()
    - 스트림 요소의 개수를 long 타입으로 리턴
  - findFirst()
    - 스트림의 첫 번째 요소를 Optional 객체에 담아 리턴
  - max(Comparator.comparing(람다식))
    - Stream<제네릭> 타입일 때만 매개변수를 넣어줌
  - min(Comparator.comparing(람다식))
    - Stream<제네릭> 타입일 때만 매개변수를 넣어줌
  - collect
    - 스트림의 요소를 수집해서 Collection 객체로 변환
    - Collectors.toMap(key 지정 람다식, value 지정 람다식)
      - collect 메소드를 이용해 자료 형태를 Map형으로 변환
    - Collectors.groupingBy(key 지정 람다식, value 지정 람다식)
      - 맵으로 묶어주는 키를 지정하는 메소드
      - 키로 지정된 값들 중 동일한 값들을 묶어서 그룹화
        - 밸류 람다식을 작성하지 않으면, 키로 지정된 요소와 같은 값인 요소들을 그룹화 하여 리스트로 만듦
      - averagingDouble(평균값을 구할 람다식)
        - 그룹화된 리스트의 평균값을 구하는 메소드
      - counting()
        - 그룹화된 리스트 안의 요소의 개수를 구하는 메소드

#### **4. 파이프라인**
- 구성 시, 최종 단계 메소드의 리턴 값이 리턴됨.
  - 최종 단계 메소드가 리턴하는 타입의 변수로 값을 받아줌.
  - ➡ 항상 최종 처리 단계를 추가할 필요가 있음.

#### **5. 스트림 배열**
- 주의 사항
  - 기본 타입 배열들은 Stream<제네릭> 형태로 만들 수 없음
  - IntStream, DoubleStream, LongStream ... 활용
  - String은 클래스이자 참조타입이므로 Stream<String> 형태로 변환

- 변환 방법
  - Arrays.stream()
  ```
  예시) IntStream intStream1 = Arrays.stream(intArr);
  ```
  - Stream.of()
  ```
  예시) IntStream intStream2 = IntStream.of(intArr);
  ```

- 포크조인 프레임워크
  - 병렬 스트림 처리를 위한 프레임워크
- parallelStream()
