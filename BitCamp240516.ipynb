### **Lambda** (교재 694p)
---
#### **1. 특징**
  - 익명함수를 간결하게 표현할 수 있는 방법
  - 프로그래밍 스타일 지원 ➡ 코드의 가독성 및 간결성을 높임

#### **2. 기본 구문**
```
LambdaEx.excuteOperation(parameters, (variables) -> {expressions};);
```
  - 변수가 없거나 2개 이상인 경우 () 생략 불가
  - 표현식이 2개 이상인 경우 {} 생략 불가
    - 무조건 {} 안에서만 세미콜론(;) 을 사용할 수 있다.

#### **3. *@FunctionalInterface***
  - 함수형 인터페이스 선언
  - 추상 메소드는 하나만 가질 수 있음
  - 컴파일러과정에서 추상메소드가 하나인지 검사

#### **4. 메소드의 참조**
- 목적
  - 메소드를 참조해 매개변수의 정보 및 리턴 타입을 알아냄
  - ➡ 람다식에서 불필요한 매개변수 제거

- 조건
  - 참조하는 메소드의 형태가 함수형 인터페이스에 선언한 추상메소드의 형태와 동일해야 함(리턴타입, 매개변수의 순서, 개수, 타입)

- 형식
  - static 메소드를 참조할 때
  ```
  Class :: Method
  ```
  - 일반(인스턴스) 메소드를 참조할 때
    - 메소드를 사용할 객체를 먼저 생성 후
    ```
    Reference :: Method
    ```

##### **4-1. 생성자 참조**
- 조건
  - 참조하는 메소드의 형태가 함수형 인터페이스에 선언한 추상메소드의 형태와 동일해야 함(리턴타입, 매개변수의 순서, 개수, 타입)
- 형식
  ```
  Class :: new
  ```


  

### **Stream** (교재 722p)
---
#### **1. 특징**
- 컬렉션 및 배열 기반
- 내부 반복
  - 처리 속도가 빠르고 병렬 처리에 효율적
  - 병렬 처리:
- 지연 평가
  - 필요한 시점에만 요소를 처리함
- 함수형 프로그래밍 스타일
  - forEach
    - 람다식으로 다양한 요소 처리를 저장할 수 있으며, 중간 처리와 최종 처리 수행을 위한 파이프라인을 형성
      - 중간 처리
        - 리턴 타입 Stream
        - 여러번 사용할 수 있음
      - 최종 처리
        - 리턴 타입 Stream 아님
        - 마지막에 한 번만 사용할 수 있음
  - 요소 필터링, 변환 등의 연산을 함수 형태로 제공
  - ➡ 메소드 호출 시, 람다식으로 함수형 인터페이스의 구현체 전달

#### **2. 기본 구문**
- 컬렉션
```
List<타입> 리스트명 = Array.asList(변수);
Stream<타입> 스트림명 = 리스트명.stream();
```
- 배열
```
타입[] 배열명 = {배열 값};
Stream<타입> 스트림명 = Arrays.stream(배열);
```
  - 기본 타입의 배열들을 스트림으로 변환하여 사용할 수 있는 기본 타입 스트림 존재
  - IntStream, LongStream, DoubleStream ...

#### **3. 응용 함수**
  - mapToInt()
    - int만 모여 있는 스트림으로 변환
  - sum()
    - 내부의 모든 값들을 합산함
  - filter()
    - true나 false가 리턴되도록 논리식이나 조건식을 사용

#### **4. 파이프라인**
- 구성 시, 최종 단계 메소드의 리턴 값이 리턴됨.
  - 최종 단계 메소드가 리턴하는 타입의 변수로 값을 받아줌.
  - ➡ 항상 최종 처리 단계를 추가할 필요가 있음.
