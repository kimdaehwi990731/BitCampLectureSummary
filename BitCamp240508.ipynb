#### **String**
---
- 특성
  - 참조 타입: 객체를 만들어 사용
  - 불변성
    - ① 처음 생성된 문자열이 메모리에 저장
    - ② 같은 변수에 다른 문자열을 대입할 때, 메모리에 저장된 값이 변경 x, 새로운 문자열을 만들어 메모리에 저장
    - StringBuilder: 메모리에 저장된 문자열의 값을 수정할 수 있음.

- 활용되는 메소드

```
String str1 = "NAVER";
String str2 = "CLOUD";
String str3 = "naver";

// 1. equals 메소드를 사용하여 대소문자를 구분하여 문자열 비교
if (str1.equals(str3)) {
    System.out.println("같은 문자열입니다.");
} else {
    System.out.println("다른 문자열입니다."); // 출력
}

// 2. equalsIgnoreCase 메소드를 사용하여 대소문자를 구분하지 않고 문자열 비교
if (str1.equalsIgnoreCase(str3)) {
    System.out.println("같은 문자열입니다."); // 출력
} else {
    System.out.println("다른 문자열입니다.");
}

// 3. compareTo 메소드를 사용하여 대소문자를 구분하여 문자열 비교
System.out.println(str1.compareTo(str2)); // 대소문자 구분하여 다른 문자열이므로 음수 출력
System.out.println(str1.compareTo(str3)); // 대소문자 구분하여 다른 문자열이므로 양수 출력

// 4. concat 메소드를 사용하여 두 문자열을 합쳐서 새로운 문자열 반환
String newStr = str1.concat(str2);
System.out.println(newStr); // 출력값: "NAVERCLOUD"
```
```
String str1 = "computer";
String str2 = "smartPhone";
String str3 = "abcdha bcdeba";

// 5. contains: 문자열을 포함하는지 여부를 검사하는 메소드
// 포함하고 있으면 true, 포함하지 않으면 false 반환
System.out.println(str2.contains("tP")); // 출력값: false
System.out.println(str3.contains("tP")); // 출력값: false

// 6. indexOf, lastIndexOf: 특정 문자나 문자열의 index를 반환하는 메소드
// 포함되어 있으면 해당 문자의 인덱스를 찾아서 반환, 포함되어 있지 않으면 -1 반환
// indexOf는 왼쪽에서부터 오른쪽으로, lastIndexOf는 오른쪽에서부터 왼쪽으로 검사
System.out.println("str2에서 h의 index: " + str2.indexOf('h')); // 출력값: 3

System.out.println("str3에서 a의 index: " + str3.indexOf('a')); // 출력값: 0
System.out.println("str3에서 index 4부터 a의 index: " + str3.indexOf('a', 4)); // 출력값: 10

System.out.println("str3에서 abc의 index: " + str3.indexOf("abc")); // 출력값: -1
System.out.println("str2에서 abc의 index: " + str2.indexOf("abc")); // 출력값: -1

System.out.println("str3에서 마지막 b의 index: " + str3.lastIndexOf('b')); // 출력값: 11

// 7. length: 문자열의 길이를 반환하는 메소드
System.out.println("str3의 길이: " + str3.length()); // 출력값: 14

// 배열에서는 length가 메소드가 아니고 속성값(멤버 변수)임
int[] intArr = new int[100];
System.out.println("intArr의 길이: " + intArr.length); // 출력값: 100

// 8. List의 길이는 size 메소드 사용
List<Integer> intList = new ArrayList<>();
intList.add(1);
intList.add(2);
intList.add(3);

System.out.println("intList의 길이: " + intList.size()); // 출력값: 3

// 9. replace, replaceAll: 문자열에서 주어진 첫번째 매개변수를 찾아서 두번째 매개변수로 치환하는 메소드
// replace와 replaceAll의 차이
// 	- replaceAll은 정규표현식을 매개면수로 전달할 수 있음
String str4 = "abcabcabcaabbcc";

System.out.println(str4.replace("abc", "111")); // 출력값: 1111111111aabbcc
System.out.println(str4.replaceAll("abc", "111")); // 출력값: 1111111111aabbcc

// replaceAll에서 정규표현식 사용
// str4에 있는 a, c를 2로 치환
System.out.println(str4.replaceAll("[ac]", "2")); // 출력값: 2b2b22b22bb22

// 10. 체이닝 기법
// 메소드를 체인처럼 엮어서 사용하는 것
// 사용한 메소드의 리턴 타입에 따라서 엮어서 사용할 수 있는 메소드
System.out.println(str4.replace('a', '1')
                        .concat("abcabc")
                        .replace('b', '2')
                        .toUpperCase()
                    ); // 출력값: 1111111111ABCABC

```
```
String str1 = "bit-naver-cloud";
String str2 = "kiaCar";
String str3 = "aBcDeF";

// 11. split: 매개변수로 주어진 문자열로 기존 문자열을 분리해서 문자열 배열로 반환
// {"bit", "naver", "cloud"}
String[] strArr = str1.split("-");
for (String s : strArr) {
    System.out.println(s);
}

// 12. subString: 주어진 인덱스로 잘려진 문자열을 반환
// 매개변수가 하나일 때는 주어진 매개변수부터 마지막 인덱스의 문자까지 잘라진 문자열 반환
System.out.println(str2.substring(4)); // 출력값: Car

// 매개변수가 두 개일 때는 첫번째 매개변수부터 두 번째 매개변수의 인덱스까지의 문자열을 잘라서 반환
// 두번째 매개변수의 인덱스 문자는 포함하지 않는다.
System.out.println(str2.substring(2, 4)); // 출력값: aC

// 13. toLowerCase: 영문자를 모두 소문자로 변환하여 반환
// toUpperCase: 영문자를 모두 대문자로 변환하여 반환
System.out.println(str3.toLowerCase()); // 출력값: abcdef
System.out.println(str3.toUpperCase()); // 출력값: ABCDEF

// 14. trim: 문자열 앞뒤의 공백을 모두 제거하여 반환하는 메소드(공백의 크기 상관없이 모두 제거)
// 중간에 포함된 공백은 제거하지 못한다.
String str4 = "              aa     bb cc  ";
System.out.println(str4.trim()); // 출력값: aa     bb cc

// 15. valueOf: 다른 타입의 값을 String 타입으로 변경
System.out.println(String.valueOf(num1).getClass().getSimpleName()); // 출력값: String
System.out.println(String.valueOf(a).getClass().getSimpleName()); // 출력값: String
System.out.println(String.valueOf(num2).getClass().getSimpleName()); // 출력값: String

```


#### **StringBuffer**
---
- 특성
  - 불변성의 부재: 메모리에 저장돼 있는 값을 직접 변경
  - 생성
    - 크기를 지정하지 않으면 Buffer의 크기가 16으로 저장됨
- 활용되는 메소드

```
// 1. StringBuffer 객체 생성
// buffer 크기가 16인 StringBuffer 객체 생성
StringBuffer sb1 = new StringBuffer();

// 버퍼 크기가 32인 StringBuffer 객체 생성
StringBuffer sb2 = new StringBuffer(32);

// bitcamp라는 문자열 값을 갖는 StringBuffer 객체 생성
StringBuffer sb3 = new StringBuffer("bitcamp");
StringBuffer sb4 = new StringBuffer("bitcamp");

// capacity(): StringBuffer에 버퍼의 크기를 리턴하는 메소드
System.out.println(sb1.capacity()); // 출력값: 16
System.out.println(sb2.capacity()); // 출력값: 32
System.out.println(sb3.capacity()); // 출력값: 23 (기본 버퍼 크기 16 + 문자열 길이 7)

// length(): StringBuffer에 저장된 문자열의 길이를 리턴하는 메소드
System.out.println(sb3.length()); // 출력값: 7

// 2. StringBuffer에서는 equals 메소드가 오버라이드 되어 있지 않음
// --> 같은 객체인지를 비교한다.
System.out.println(sb3.equals(sb4)); // 출력값: false (같은 문자열이라도 false로 출력됨)

// StringBuffer의 문자열 값을 비교하는 방법: String 타입으로 변환 후 equals 메소드 사용
// StringBuffer를 String 타입으로 변환하는 방법: toString() 메소드 활용 < 이게 더 깔끔하다
System.out.println(sb3.toString().equals(sb4.toString())); // 출력값: true
System.out.println(String.valueOf(sb3).equals(String.valueOf(sb4))); // 출력값: true
```

```
// 1. append: 기존 StringBuffer에 새로운 문자열을 추가하여 StringBuffer 타입으로 반환
// StringBuffer의 append는 원본 데이터를 수정한다.
// sb1에 저장돼 있던 문자열이 수정됨
sb1.append("navercloud");
System.out.println(sb1); // 출력값: navercloud

sb3.append("bitcamp");
System.out.println(sb3); // 출력값: bitcamp

// String은 원본 데이터는 수정되지 않는다.
// str에 저장돼 있는 문자열은 그대로 유지됨.
String str = "a";
str.concat("navercloud");
System.out.println(str); // 출력값: a

// 2. capacity: 버퍼의 크기 반환
//    버퍼의 크기를 지정한 것보다 길이가 긴 문자열을 append하면 자동으로 문자열의 길이 맞춰 버퍼의 크기가 증가한다
//    length: StringBuffer에 저장된 문자열의 길이 반환
System.out.println(sb1.capacity()); // 출력값: 108 (기본 크기 16 + "navercloud"의 길이 10)
System.out.println(sb1.length()); // 출력값: 10 (문자열 "navercloud"의 길이)

//test
StringBuffer aaa = new StringBuffer(8);
// length가 18이니깡 ( 기본 기준: 16이하일때는 16으로 고정, 넘어가는 순간 16 + 문자열길이)
// 지정된 수용량 이하까지는 그걸로 고정 그거 넘어가는 순간 기존 + 문자열길이

// 3. StringBuffer에 남아 있는 버퍼의 크기
aaa.append("000011112222");

int remainBufferSize = aaa.capacity() - aaa.length();
// 남은 것만큼 글자 채워주기
for (int i = 0; i < remainBufferSize; i++) {
    aaa.append("3");
}
System.out.println(aaa); // 출력값: 0000111122223333
aaa.append("A");
System.out.println("aaa 크기" + aaa.capacity()); // 출력값: aaa 크기26

// 4. delete: 매개변수로 주어진 시작 인덱스부터 끝 인덱스까지의 문자열을 삭제한 새로운 StringBuffer 반환
sb2.delete(3, 7);
System.out.println(sb2); // 출력값: bitp

// deleteCharAt: 해당 인덱스의 문자 하나만 삭제
sb2.append("camp")
    .deleteCharAt(0);
System.out.println(sb2); // 출력값: itp

// 5. insert: 첫번째 매개변수 위치에 두번째 값을 삽입
sb2.insert(0, "b");
System.out.println(sb2); // 출력값: bitp

// 6. reverse: StringBuffer에 저장된 문자열을 역순으로 정렬
sb2.reverse();
System.out.println(sb2); // 출력값: ptib

// 7. setCharAt: 첫번째 매개변수로 주어진 인덱스 위치에 있는 문자를 두번째 매개변수로 주어진 값으로 수정하여 반환
sb2.setCharAt(0, 'a');
System.out.println(sb2); // 출력값: atib

// 8. setLength: STringBuffer에 저장되어 있는 문자열의길이를 매개변수로 변경
// 저장돼 있는 길이보다 매개변수가 클 때는 빈 곳을 공백으로 채운다.
// 저장돼 있는길이보다 매개변수가 작을 때는 저장돼 있는 문자열에서 남은 부분 삭제.
// 버퍼의 크기는 변경하지 못한다
sb2.setLength(20);
sb1.setLength(20);
System.out.println(sb2); // 출력값: atib
System.out.println(sb1); // 출력값: navercloud         
// sb2에 있는 공백삭제
System.out.println(sb2.toString().trim()); // 출력값: atib
```

#### **Math**
---
- 활용되는 메소드

```
// 1. 소수점 첫째 자리에서 올림
System.out.println("1. Math.ceil(num): " + Math.ceil(1234.5678)); // 출력값: 1235.0

// 2. 소수점 셋째 자리까지 올림
System.out.println("2. Math.ceil(num * 1000) / 1000: " + Math.ceil(1234.5678 * 1000) / 1000); // 출력값: 1234.568

// 3. 소수점 첫째 자리에서 반올림
System.out.println("3. Math.round(num): " + Math.round(1234.5678)); // 출력값: 1235

// 4. 소수점 셋째 자리까지 반올림
// long 타입으로 반환되기 때문에 원하는 값을 얻지 못함
// double 타입으로 변환 후 계산해야 함
System.out.println("4. (double)Math.round(num * 1000) / 1000: " + (double)Math.round(1234.5678 * 1000) / 1000); // 출력값: 1234.568

// 5. 소수점 첫째 자리에서 버림
System.out.println("5. Math.floor(num): " + Math.floor(1234.5678)); // 출력값: 1234.0

// 6. 소수점 셋째 자리까지 버림
System.out.println("6. Math.floor(num * 1000) / 1000: " + Math.floor(1234.5678 * 1000) / 1000); // 출력값: 1234.567
```
```
int num1 = 123;
int num2 = -456;

// 1. abs: 절대값을 반환하는 메소드
System.out.println(Math.abs(num1)); // 출력값: 123
System.out.println(Math.abs(num2)); // 출력값: 456

// 2. max: 두 개의 매개변수 중 큰 값을 반환
//    min: 두 개의 매개변수 중 작은 값을 반환
//    두 개의 매개변수가 동일한 타입으로 전달돼야 한다.
System.out.println(Math.max(num1, num2)); // 출력값: 123
System.out.println(Math.min(num1, num2)); // 출력값: -456

// 3. rint: 매개변수로 주어진 실수에서 가장 가까운 정수값을 double 타입으로 반환(근사치)
System.out.println(Math.rint(123.456)); // 출력값: 123.0
System.out.println(Math.rint(1.5)); // 출력값: 2.0
System.out.println(Math.rint(2.5)); // 출력값: 2.0
```


#### **Wrapper**
---

```
Integer num1 = 10;
Integer num2 = Integer.valueOf(100);
Integer num3 = Integer.valueOf("100");

// 2. 기본 타입은 값을 저장하는 기능 제공
// 속성, 메소드를 사용할 수 없음
// 속성, 메소드 사용을 위해선 기본 타입에 매핑된 래퍼클래스 객체로 변환해서 사용해야 함
int num4 = 30;

Integer num5 = Integer.valueOf(num4);

Integer num6 = Integer.valueOf(100);
// 3. 같은 객체를 사용하는지 비교
// valueOf 메소드로 생성되는 객체는 갑이 같으면 같은 객체를 사용한다.
// -128 ~ 127까지만 같은 객체를 반환하고, 범위를 벗어난 값은 새로운 객체를 생성
System.out.println(num2 == num6); // false

// 4. equals 메소드는 오버라이드가 돼 있어 값의 동일여부를 비교한다.
System.out.println(num1.equals(num6));
System.out.println(num1.equals(num2));

// 5. comparTo: 객체에 저장돼 있는 값이 같으면 0, 다르면 1, -1 리턴(앞에 온 것이 매개변수보다 작으면 -1, 크면 1)
Integer num7 = Integer.valueOf(5);	// integer 타입의 객체 리턴

System.out.println(num1.compareTo(num6));
System.out.println(num1.compareTo(num7));

// 6. parseInt: 문자열 값을 int 타입으로 변환
String numStr = "1000";
int convertNum = Integer.parseInt(numStr);	// int 타입으로 리턴
long convertLongNum = Long.parseLong(numStr);

// 7. int 타입의 최대값과 최소값을 상수(static final)로 지정해놔서
// 바로 사용할 수 있다.
System.out.println(Integer.MAX_VALUE);
System.out.println(Integer.MIN_VALUE);

// 8. toString: Integer객체를 String 타입의 객체로 변환
String convertString = num1.toString();
System.out.println(convertString.getClass().getSimpleName());

// 9. 생성된 Integer 객체의 사이즈, 바이트, 타입 확인
System.out.println(num1.SIZE);
System.out.println(num1.BYTES);
System.out.println(num1.TYPE);
```

```
int num1 = 100;

// 1. 박싱(boxing): 기본 타입의 값을 래퍼클래스의 객체형태로 포장해주는 작업
Integer objNum1 = (Integer)num1;

// 2. 자동 박싱: 컴파일러가 알아서 기본 타입의 값을 래퍼클래스의 객체로 변환하는 기능
Integer objNum2 = objNum1 + num1;

// 3. 언박싱(unboxing): 래퍼클래스 객체에 저장되어 있는 값을 포장을 해체하여 기본타입의 값으로 변환하는 작업
int num2 = (int)objNum2;

// 4. 자동 언박싱: 컴파일러가 알아서 래퍼클래스의 객체에 저장되어 있는 값을 기본타입 값으로 변환해주는 기능
int num3 = num2 + objNum2;

// 5. 메소드에서의 박싱, 언박싱
printObjNum(1000);

printPrimitiveNum(objNum2);

// 6. 제네릭에서의 박싱
List<Integer> intList = new ArrayList<>();

intList.add(30);

System.out.println(intList.get(0).getClass().getSimpleName());

public static void printObjNum(Integer integer) {
  System.out.println(integer);
}

public static void printPrimitiveNum(int intVal) {
  System.out.println(intVal);
}
```
